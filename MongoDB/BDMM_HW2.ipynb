### Question 3

Once per month we like to reward hosts with recognition. Pick three superhosts who have at least two property listings that can accommodate more than four people?


```python
#We have some specifications that need to be met regarding some field values. We host needs to be a superhost, so we only
#want documents with superhost field as 1. The property needs at lest 2 listings, so again we only want documents with
#host listings count bigger or equal than 2, so we use gte (that returns all documents with value equal or greater that the
#one specified), and inally properties that can accomodate more than 4 people, meaning 5 or more people.
query_1 = {
    '$match': {
        'host_is_superhost': True,
        'host_listings_count': {'$gte':2},
        'accommodates' : {'$gte': 5}
        
    }
}

#Since we only want three superhosts, we limit the output to 3.
query_2 = {
    '$limit': 3
}

pipeline = [query_1, query_2]
iterator = db.listingsAndReviews_new.aggregate(pipeline)
result = list(iterator)

pprint('Superhosts with specified characteristics: ' + str(len(result)))
pprint(result)
```

### Question 5

We are thinking of identifying someone to hire to write review for us professionally. What is the name of the reviewer who left the longest review in New York?

```python
#Set the location as New York
location_id = 'New York, New York, United States'

#Check only the transactions where the host location matches the desired location.
query_1 = {
    '$match': {
        'host_location': location_id 
    }
}

#Unwind the reviwes so we have one review per document, making the reviews properties accessible to us.
query_2 = {
        '$unwind': '$reviews'
}

#Limit the number of documents to return, just for exercise preparation reasons, we should remove these, but first we have
#to see how to drop missing values from the collection.
query_3 = {
    '$limit': 10
}

#We only want to return the reviewer id. We will also return the length of the comment, so we can check if the output is
#correct.
query_4 = {
    '$project': {
        'username': 1,
        'reviewer': '$reviews.reviewer_name',
        'length': { '$strLenCP': '$reviews.comments'}
    }
}

#To obtain the max length, we order the documents per comment length in descending order.
query_5 = {
    '$sort': {'length':-1}
}

#Finally, get the first document, the reviewer who left the longest review in New York.
query_6 = {
    '$limit': 1
}

pipeline = [query_1, query_2, query_3, query_4, query_5, query_6]
iterator = db.listingsAndReviews_new.aggregate(pipeline)
result = list(iterator)

pprint('Reviewer who left the  longest review in New York: ' + str(len(result)))
pprint(result)
```

### Question 7

We want to identify areas by whether they are typically used for short breaks, like weekend mini breaks, or whether they are more suitable for long trips. We will use this information to target advertising of different customers. It is not expected this information will change much over time so we won’t look to update it we just would like a current view. What is the average duration of stay (in nights) per type of property per city (you can use the maximum_nights to measure length of stays)? For each property type return the city with the highest and lowest average value.

```python
#We understood that we should return the city with the highes and the city with the lowest average values per property type.

#Group the average duration (using maximum_nights) by property type and city (we used address.market since it was the field
#that we understtod that best represented the city of the property).
query_1 = {
    '$group': {
        '_id': {'property':'$property_type', 'city':'$address.market'},
        'Average_duration': {'$avg': {'$toInt': '$maximum_nights'}}
    }
}

#Group by property type and create an array named city that stores both city name and the average duration for that city.
#With the push command we create that array and assign values to the elements. With this we are able to compute the average
#for all cities per property type without having a complex primary key. Here the primary key is only the property type, and
#we can calculate the average duration of every city that has that type of property.
query_2 = {
    '$group': {
        '_id': '$_id.property',
        'cities': {
            '$push': {
                'city': '$_id.city',
                'Average_duration': '$Average_duration'
            }
        },
        #Then we check the city with the maximum average, keeping that value and the name of the city, per property type.
        'City_max': {'$max': {
            'city': '$_id.city',
            'Average_duration': '$Average_duration'
            }
        },
        #We also check the city with the minimum average, keeping also that value and the name of the city, per property
        #type.
        'City_min': {'$min': {
            'city': '$_id.city',
            'Average_duration': '$Average_duration'
            }
        }
    }
}

#Finally create a query projection that will only display the property type and the city with higher average duration and
#the city with the lowest average duration for each property type
query_3 = {
    '$project': {
        'username': 1,
        'City max': '$City_max.city',
        'Max average': '$City_max.Average_duration',
        'City min': '$City_min.city',
        'Min duration': '$City_min.Average_duration'
    }
}


pipeline = [query_1, query_2, query_3]
iterator = db.listingsAndReviews_new.aggregate(pipeline)
result = list(iterator)

pprint('Max and Min average duration of stay: ' + str(len(result)))
pprint(result)
```
